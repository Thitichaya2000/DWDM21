##0. หารัฐที่มีจำนวนเครื่องบินขาเข้ามากที่สุดและน้อยที่สุด

````python
merged_table = flights.merge(airports,how='left',left_on='รหัสสนามบินปลายทาง',right_on='รหัสสนามบิน') # how = ตารางหลัก left_on ตารางซ้ายของตารางหลัก ผูกกับ right_on id
merged_table
````

````python
merged_table.dropna() ##เช็ค missing
````

````python
merged_table.isnull() .any() #เช็คดู ถ้ามี true แสดงว่ามีค่า missing
````

````python
merged_table.groupby('รัฐ').count() #ตรวจดูจำนวนเที่ยวบินของแต่ละรัฐ
````

````python
max = merged_table.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False) #รัฐที่มีจำนวนเครื่องบินขาเข้ามากที่สุด
max.iloc[:1,4:5]
````

````python
min = merged_table.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False)  #รัฐที่มีจำนวนเครื่องบินขาเข้าน้อยที่สุด
min.iloc[-1:,4:5]
````

##5.วันธรรมดาหรือวันเสาร์-อาทิตย์มีเครื่องบินขาเข้ามากกว่ากัน (วันจันทร์ = 1, วันอังคาร = 2, ...)


````python
merged_table.groupby('วันในสัปดาห์').count()
````

````python
week_Arrival_plane = merged_table.groupby('วันในสัปดาห์').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False) #ดูจำนวนเครื่องบินขาเข้าในแต่ละวัน
week_Arrival_plane
````

````python
Arrival_on_weekdays = week_Arrival_plane.iloc[:5,3:4].sum()  #หาจำนวนเครื่องบินขาเข้าในวันธรรมดา
Arrival_on_weekdays
````

````python
Holiday_arrivals = week_Arrival_plane.iloc[5:7,3:4].sum() #หาจำนวนเครื่องบินขาเข้าในวัน เสาร์-อาทิตย์
Holiday_arrivals
````

วันธรรมดามีเครื่องบินขาเข้า  2011453

วันเสาร์-อาทิตย์มีเครื่องบินขาเข้า 690765

**ดังนั้น วันธรรมดามีเครื่องบินขาเข้ามากกว่าวันเสาร์-อาทิตย์**

##6.เรียงลำดับ ชื่อสนามบิน ที่มีรหัสสนามบิน 5 อันดับแรก

````python
merged_table = flights.merge(airports,how='left',left_on='รหัสสนามบินปลายทาง',right_on='รหัสสนามบิน') # how = ตารางหลัก left_on ตารางซ้ายของตารางหลัก ผูกกับ right_on id
merged_table
````

````python
merged_table.groupby('ชื่อสนามบิน').count() 
````

````python
airport_name =merged_table.groupby('ชื่อสนามบิน').count().sort_values(by=['รหัสสนามบิน'],ascending=False)
airport_name 
````

````python
#เรียงลำดับ ชื่อสนามบิน ที่มีรหัสสนามบิน 5 อันดับแรก
airport_name_cut  = airport_name .iloc[:5,7:8] 
airport_name_cut
````
